## Linear Regression GridSearch##

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, ElasticNetCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
# Firstly linear regression with regularisation (gridsearched)
```

```python
LFB = pd.read_csv('../Assets/LFBgts.csv')
```

```python
LFB.info()
```

```python
LFB.ob_time = pd.to_datetime(LFB.ob_time)
LFB.set_index('ob_time', inplace = True, drop=True)
LFB=LFB.sort_index(ascending = True)
```

```python
LFB.head()
```

```python
LFB1 = LFB[[
 'HourOfCall', 'callouts', 'cld_ttl_amt_id', 'islight', 'max_gust_speed', 'prcp_amt', 'rltv_hum', 'stn_pres', 'visibility',
 'wind_speed', 'wmo_hr_sun_dur', 'holiday', 'weekend', 'air_temperature', 'prioryear', 'wind_dirc', 'lockdown', 'tempsq']]
```

```python
LFBdum = pd.get_dummies(
    LFB1, columns=['wind_dirc', 'HourOfCall'])
```

```python
y = LFBdum.pop('callouts')
X = LFBdum.copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# create an ElastinNetCV model instance
modellrgs = ElasticNetCV(alphas=np.logspace(-4, 1, 30), 
                     l1_ratio=np.array([ 0.75, 0.775,0.8,0.825,0.85,0.875,0.9,0.95,1]),
                     cv=5, max_iter=1000000,verbose =1,n_jobs=-2)
# fit the model
modellrgs.fit(X_train, y_train)
# get the best alpha
print('Best alpha:', modellrgs.alpha_)
# get the best l1-ratio
print('Best l1-ratio:', modellrgs.l1_ratio_)
# evaluate on the training set
print('Training score:', modellrgs.score(X_train, y_train))
# evaluate on the test set
print("Test Score:", modellrgs.score(X_test, y_test))
print('CV score: ', (cross_val_score(modellrgs, X_train, y_train)).mean())
```

```python
# Plotting the model coefficients.
plt.figure(figsize = (20,20))
# collect the model coefficients in a dataframe
df_coef = pd.DataFrame(modellrgs.coef_, index=LFBdum.columns,
                       columns=['coefficients'])

# plot the magnitude of the coefficients
df_coef['coefficients'].sort_values().plot(kind='barh')
plt.show()
```

```python
# Plotting predicted vs actual.

predictions = modellrgs.predict(X_test)

#residuals (or error between predictions and actual)
residuals = y_test - predictions

# Plot predicted vs actual 
plt.figure(figsize=(8,8))
plt.scatter(predictions, y_test, s=30, c='r', marker='+', zorder=10)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='k', lw=2)
plt.xlabel("Predicted Values - $\hat{y}$")
plt.ylabel("Actual Values callouts - y")
plt.show()
```

```python
# Plot standardised residuals
predictionsall = modellrgs.predict(X)
residualsall = y - predictionsall

stdres = (residualsall - residualsall.mean())/residualsall.std()
plt.figure(figsize=(20,20))
plt.plot(X.index, stdres,  c='r', marker='+', zorder=10)
#plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='k', lw=2)
plt.xlabel("Index")
plt.xticks(rotation = '45')
plt.ylabel("Standardised Residuals")
plt.show()
```

```python
# Investigating standardised residuals distribution.
import scipy
fig = plt.figure(figsize=(15,10))
sns.distplot(stdres,fit=scipy.stats.norm, kde=True, hist=True, color='r')
plt.show()
```

```python
# QQplot
import statsmodels.api as sm 
import pylab as pyl

sm.qqplot(stdres, line ='45') 
pyl.show() 
```

```python
# Creating model for use in the prediction package later.
alpha = modellrgs.alpha_ 
l1_ratio = modellrgs.l1_ratio_
# create an ElasticNet model instance
modellr = ElasticNet(alpha=alpha, l1_ratio=l1_ratio,max_iter=100000)

# get cross validated scores
scores = cross_val_score(modellr, X_train, y_train, cv=5)
print("Cross-validated training scores:", scores)
print("Mean cross-validated training score:", scores.mean())
# fit and evaluate the data on the whole training set
modellr.fit(X_train, y_train)
print("Training Score:", modellr.score(X_train, y_train))
# evaluate the data on the test set
print("Test Score:", modellr.score(X_test, y_test))
```
