```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# set the filepath and load in a shapefile
fp = "Assets/London_Borough_Excluding_MHW.shp"

map_df = gpd.read_file(fp)
```

```python
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
map_df.head()
```

```python
map_df.NAME = map_df.NAME.str.upper()
```

```python
# now let's preview what our map looks like with no data in it
map_df.plot()
```

```python
# load in new csv file
df = pd.read_csv("Assets/LFBIncident2017.csv", header=0)

df.head()
```

```python
list(df.columns)
```

```python
df1 = df[df['CalYear'] == 2017][['IncidentNumber', 'IncGeo_BoroughName']]
```

```python
#titanic.groupby(['Sex', 'Pclass'], as_index=False)[
#    ['Survived', 'Age', 'Fare']].mean()

df2 =df1.groupby('IncGeo_BoroughName', as_index = False)['IncidentNumber'].count()
df2.head()
```

```python
score = df2.rename(index=str, columns={"IncGeo_BoroughName": "borough",
                                      "IncidentNumber": "2020callouts"
                                      })
```

```python
merged = map_df.set_index('NAME').join(score.set_index('borough'))

merged.head()
```

```python
merged['2020callouts'].value_counts()
```

```python
merged.isna().sum()
```

```python
variable = '2020callouts'

# set the range for the choropleth
vmin, vmax = 1220, 4600

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 6))

# create map
merged.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')

# Now we can customise and add annotations

# remove the axis
ax.axis('off')

# add a title
ax.set_title('LFB 2020 Callouts By Borough', \
              fontdict={'fontsize': '25',
                        'fontweight' : '3'})

# create an annotation for the  data source
ax.annotate('Source: London Datastore, 2014',
           xy=(0.1, .08), xycoords='figure fraction',
           horizontalalignment='left', verticalalignment='top',
           fontsize=10, color='#555555')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)

# this will save the figure as a high-res png. you can also save as svg
fig.savefig('Assets/images/calloutsmap.png', dpi=300)
```

```python
background = merged.to_crs(epsg = 3857)
```

```python
ax = background.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
ctx.add_basemap(ax)
plt.show()
```
