# Classifier Modelling #

```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, mean_squared_error,r2_score
from sklearn.metrics import confusion_matrix, accuracy_score, plot_confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import scikitplot as skplt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.color_palette("husl", 3))
```

## Loading the data ##

```python
LFBclass = pd.read_csv('../Assets/LFBClass.csv')
```

```python
LFBclass.info()
```

```python
LFBclass['ob_time'] = pd.to_datetime(LFBclass.ob_time)
LFBclass.set_index('ob_time', inplace=True, drop=True)
```

```python
y = LFBclass.pop('IncidentGroup')
```

## Extracting the Baseline ##

```python
y.value_counts(normalize = True)
```

```python
y.value_counts()
```

```python
dummify = ['wind_dirc','PropertyType', 'HourOfCall']
X = pd.get_dummies(LFBclass, columns=dummify, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## First classifier - Logistic Regression. ##

```python
lr = LogisticRegression(multi_class='multinomial',max_iter = 10000)
cross_val_score(lr, X_train, y_train, cv=5,n_jobs = -2).mean()
```

```python
lr.fit(X_train, y_train)
predictionslr = lr.predict(X_test)
```

```python
fig, ax = plt.subplots()
metrics.plot_confusion_matrix(lr, X_test, y_test, labels=['Special Service', 'False Alarm','Fire'], cmap='Blues', ax=ax)

texts = ax.texts
for text in texts:
    text.set_size(16)

plt.show()
```

```python
print(metrics.classification_report(y_test, predictionslr, digits=4))
```

```python
probabilitieslr = lr.predict_proba(X_test)
```

```python
skplt.metrics.plot_precision_recall(y_test, probabilitieslr, 
                       plot_micro=True,
                       title_fontsize=20, text_fontsize=16, figsize=(8,6),cmap=cmap)
plt.show()
```

```python
skplt.metrics.plot_roc(y_test, probabilitieslr, plot_micro=True, plot_macro=True, 
                       title_fontsize=20, text_fontsize=16, figsize=(8,6), cmap=cmap)
plt.show()
```

```python
base_estimator = DecisionTreeClassifier(max_depth=3, max_features=0.8)
decbag = BaggingClassifier(base_estimator=base_estimator,
                           n_estimators=1000,
                           random_state=1)
```
